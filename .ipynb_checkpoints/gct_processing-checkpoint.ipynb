```python
import pandas as pd
```

```python
#1.
def read_gct(filepath: str, summary: bool = False) -> pd.DataFrame:
    # reads the GenePattern GCT file, skipping the metadata headers
    df = pd.read_csv(filepath, sep='\t',skiprows=2)
    
    #set name & description as indexes separate from the data
    df.set_index(["Name","Description"], inplace=True)
    
    #prints num of rows & columns if summary is requested
    if summary:
        print(f"Imported {df.shape[0]} rows and {df.shape[1]} columns.")
        
    return df
```

```python
#2a.
file = 'https://datasets.genepattern.org/data/test_data/BRCA_minimal_60x19.gct'
df = read_gct(file, True)
```

```python
#2b.
file2 = 'https://datasets.genepattern.org/data/test_data/BRCA_large_20783x40.gct'
df = read_gct(file2)
```

```python
import matplotlib.pyplot as plt
```

```python
#3.
#array of sample names
samples = ["A7-A0DB-normal", "A7-A13E-normal", "BH-A0B3-primary", "BH-A0B5-primary"]

#plot a histogram for each sample
for sample in samples:
    plt.hist(df[sample], bins=50, range=(0,50000), edgecolor='black')
    plt.title(f"{sample}")
    plt.xlabel("Expression Value of Gene")
    plt.ylabel("Frequency of Expression Value")
    plt.show()
```

```python
#4.
df_stats = pd.DataFrame({
    "Sample name": df.columns,
    "Mean": df.mean(axis=0).values,
    "Median": df.median(axis=0).values,
    "Standard deviation": df.std(axis=0).values
})

df_stats
```

```python
#5a.
filtered_df = df[df.max(axis=1) >= 1000]
filtered_df
```

```python
#5b.
#means & medians of each column in filtered dataframe
filtered_stats = pd.DataFrame({
    "Sample name": filtered_df.columns,
    "Mean": filtered_df.mean(axis=0).values,
    "Median": filtered_df.median(axis=0).values
})
filtered_stats

#plot 4 histograms again
samples = ["A7-A0DB-normal", "A7-A13E-normal", "BH-A0B3-primary", "BH-A0B5-primary"]
for sample in samples:
    plt.hist(filtered_df[sample], bins=50, range=(0, 50000), edgecolor='black')
    plt.title(f"{sample} (Filtered)")
    plt.xlabel("Expression Value of Gene")
    plt.ylabel("Frequency of Expression Value")
    plt.show()
```
